# Data Exploration

```python
import os
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from matplotlib.patches import Patch
```

```python
def colorize_image_by_labels(image, label):
    image = np.repeat(
        np.expand_dims((image / image.max()), axis=2),
        repeats=3,
        axis=2
    )
    image = (image * 255).astype(np.uint8)

    image[label==1] = [255, 0, 0]
    image[label==2] = [0, 255, 0]
    image[label==4] = [0, 0, 255]

    return image    

def grid_plot(images):
    fig = plt.figure(constrained_layout=True, tight_layout=True, figsize=(23, 20))
    grid = GridSpec(nrows=4, ncols=3, width_ratios=[0.73, 1, 1])

    legend_elements = [
                    Patch(facecolor='red', edgecolor='black', label='Necrotic Tumor Core'),
                    Patch(facecolor='lime', edgecolor='black', label='Peritumoral Edematous/Invaded Tissue'),
                    Patch(facecolor='blue', edgecolor='black', label='Gd-Enhancing Tumor')
    ]

    axis = fig.add_subplot(grid[0, 0])
    axis.imshow(images["t1ce"][0], cmap="gray")
    axis.set_yticks([])
    axis.set_xticks([])
    axis.set_title('Tranverse Image', fontdict={"fontsize": 16})
    axis.set_ylabel('T1Gd', fontdict={"fontsize": 16})
    axis = fig.add_subplot(grid[0, 1])
    axis.imshow(images["t1ce"][1], cmap="gray")
    axis.axis('off')
    axis.set_title('Coronal Image', fontdict={"fontsize": 16})
    axis = fig.add_subplot(grid[0, 2])
    axis.imshow(images["t1ce"][2], cmap="gray")
    axis.axis('off')
    axis.set_title('Sagittal Image', fontdict={"fontsize": 16})

    # labels
    axis = fig.add_subplot(grid[1, 0])
    axis.imshow(images["t1ce_colored"][0])
    axis.set_yticks([])
    axis.set_xticks([])
    axis.set_ylabel('T1Gd Labels', fontdict={"fontsize": 16})
    axis = fig.add_subplot(grid[1, 1])
    axis.imshow(images["t1ce_colored"][1])
    axis.axis('off')
    axis = fig.add_subplot(grid[1, 2])
    axis.imshow(images["t1ce_colored"][2])
    axis.axis('off')

    axis = fig.add_subplot(grid[2, 0])
    axis.imshow(images["flair"][0], cmap="gray")
    axis.set_yticks([])
    axis.set_xticks([])
    axis.set_ylabel('T2 Flair', fontdict={"fontsize": 16})
    axis = fig.add_subplot(grid[2, 1])
    axis.imshow(images["flair"][1], cmap="gray")
    axis.axis('off')
    axis = fig.add_subplot(grid[2, 2])
    axis.imshow(images["flair"][2], cmap="gray")
    axis.axis('off')

    # labels
    axis = fig.add_subplot(grid[3, 0])
    axis.imshow(images["flair_colored"][0])
    axis.set_yticks([])
    axis.set_xticks([])
    axis.set_ylabel('T2 Flair Labels', fontdict={"fontsize": 16})
    axis = fig.add_subplot(grid[3, 1])
    axis.imshow(images["flair_colored"][1])
    axis.axis('off')
    axis = fig.add_subplot(grid[3, 2])
    axis.imshow(images["flair_colored"][2])
    axis.axis('off')
    
    plt.subplots_adjust(wspace=0.0)
    fig.legend(handles=legend_elements, bbox_to_anchor=(0, 1, 1, 0), loc="lower center", mode=None, ncol=3, fontsize=16)

def parses_image_to_plot(subjet_folder, transverse_idx, coronal_idx, sagittal_idx, crop=False):
    image_scans = os.listdir(subjet_folder)
    images = {
        # [Transverse, Coronal, Sagittal]
        "t1ce": [],
        "flair": [],
        "label": [],
        "t1ce_colored": [],
        "flair_colored": []
    }

    for image in image_scans:
        if "t1ce" in image or "flair" in image:
            nii_data = nib.load(os.path.join(subjet_folder, image))
            volumetric_image = nii_data.get_fdata()

            # Gets axial images
            transverse_image = volumetric_image[:, :, transverse_idx].swapaxes(0, 1)
            coronal_image = volumetric_image[:, coronal_idx, :].swapaxes(0, 1)[-1::-1]
            sagittal_image = volumetric_image[sagittal_idx, :, :].swapaxes(0, 1)[-1::-1]

            if "t1ce" in image:
                images["t1ce"] = [transverse_image, coronal_image, sagittal_image]
            else:
                images["flair"] = [transverse_image, coronal_image, sagittal_image]

        elif "seg" in image:
            nii_data = nib.load(os.path.join(subjet_folder, image))
            volumetric_label = nii_data.get_fdata()
            
            transverse_label = volumetric_label[:, :, transverse_idx].swapaxes(0, 1)
            coronal_label = volumetric_label[:, coronal_idx, :].swapaxes(0, 1)[-1::-1]
            sagittal_label = volumetric_label[sagittal_idx, :, :].swapaxes(0, 1)[-1::-1]

            images["label"] = [transverse_label, coronal_label, sagittal_label]

    for idx in range(3):
        t1ce_image = images["t1ce"][idx].copy()
        flair_image = images["flair"][idx].copy()
        label_image = images["label"][idx]

        # Normalizes image, converts it to rgb and color pixels according to its labels
        t1ce_colored_image = colorize_image_by_labels(t1ce_image, label_image)
        flair_colored_image = colorize_image_by_labels(flair_image, label_image)

        images["t1ce_colored"].append(t1ce_colored_image)
        images["flair_colored"].append(flair_colored_image)

    grid_plot(images)
```

```python
# (240, 240, 155)
parses_image_to_plot(
    subjet_folder="../datasets/RSNA_ASNR_MICCAI_BraTS2021_TrainingData_16July2021/BraTS2021_00253",
    transverse_idx=69,
    coronal_idx=124,
    sagittal_idx=69
)
```
